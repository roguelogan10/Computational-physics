```python
# Libraries import cell
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Signal generation function
def triangular_signal(Min, Max, Step, A=1):

    """
    Generate a triangular signal.
    
    Parameters:
    x (numpy.ndarray): Input array of time values.
    
    Returns:
    numpy.ndarray: Triangular signal values corresponding to the input time values.
    """
    x = np.arange(Min, Max, Step)
    y = np.zeros_like(x)
    y[len(y)//8:7*len(y)//8] = A*(-1*abs(x[len(y)//8:7*len(y)//8]) + max(abs(x[len(y)//8:7*len(y)//8])))
    return x, y

def rectangular_signal(Min, Max, Step, A=1):

    """
    Generate a triangular signal.
    
    Parameters:
    x (numpy.ndarray): Input array of time values.
    
    Returns:
    numpy.ndarray: Triangular signal values corresponding to the input time values.
    """
    x = np.arange(Min, Max, Step)
    y = np.zeros_like(x)
    y[len(y)//8:7*len(y)//8] = A
    return x, y

# Discrete Fourier Transform function coefficients computation
def Ak(x, y, k):

    """
    Compute the cosine coefficients for a signal.
    """
    step = np.round((max(x) - min(x))/len(x),3)
    Ak = np.dot(y, np.cos(k*x))*step
    return 1/np.pi * Ak

def Bk(x, y, k):
    """
    Compute the sine coefficients for a signal.
    """
    step = step = np.round((max(x) - min(x))/len(x),3)
    Bk = np.dot(y, np.sin(k*x))*step
    return 1/np.pi * Bk

   
# Discrete Fourier Transform
def DFT(x,y,N):
    
    """
    Compute the Discrete Fourier Transform of a signal.
    
    Parameters:
    y (numpy.ndarray): Input signal to transform.
    N (int): Number of points in the DFT.
    
    Returns:
    numpy.ndarray: DFT of the input signal.
    """  
   
    DFT = Ak(x, y, 0)/2 
    for k in range(1,N):
        DFT += Ak(x,y,k)*np.cos(k*x)+Bk(x,y,k)*np.sin(k*x)
    return DFT

# Discrete Fourier Transform
def plot_signal_and_dft(x,y, list_N =[2,5,20,50]):
    plt.rcParams.update({'font.size': 18})
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(x, y, label='Original Signal')
    for n in list_N:
        ax.plot(x, DFT(x, y, n), label=f'DFT with N={n}')
    ax.set_xlabel('Time')
    ax.set_ylabel('Amplitude')
    ax.set_title('Discrete Fourier Transform of a Triangular Signal')
    ax.legend(loc='upper right')
    plt.grid()
    plt.show()
```

```python
x,y = triangular_signal(-np.pi, np.pi, 0.01, A=1)
plot_signal_and_dft(x,y)

step = np.round((max(x) - min(x))/len(x),3)
step
```

```python
x,y = rectangular_signal(-np.pi, np.pi, 0.01, A=1)
plot_signal_and_dft(x,y,list_N=range(12, 513, 500))
```
